```python
# Simple Async Example

import asyncio

async def fetch_data(name: str, delay: float) -> str:
    print(f"[{name}] Starting...")
    await asyncio.sleep(delay)  # Simulates I/O (network call, DB query, etc.)
    print(f"[{name}] Done!")
    return f"{name} result"

async def main():
    # Run tasks concurrently
    results = await asyncio.gather(
        fetch_data("Task A", 2),
        fetch_data("Task B", 1),
        fetch_data("Task C", 3),
    )
    print("\nAll results:", results)

await main()
```
