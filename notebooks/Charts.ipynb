```python
# Visualize API response data

import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load API response into a DataFrame
resp = requests.get("http://localhost:5000/passengers")
df_api = pd.DataFrame(resp.json())

# Two side-by-side charts
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Chart 1: Survival rate by passenger class
sns.barplot(data=df_api, x='pclass', y='survived', hue='pclass',
            palette='Purples_d', legend=False, ax=axes[0])
axes[0].set_title('Survival Rate by Passenger Class')
axes[0].set_ylabel('Survival Rate')

# Chart 2: Fare distribution by class (log scale due to wide range)
sns.boxplot(data=df_api, x='pclass', y='fare', hue='pclass',
            palette='Oranges', legend=False, ax=axes[1])
axes[1].set_title('Fare Distribution by Class')
axes[1].set_yscale('log')

plt.tight_layout()
plt.show()
```
